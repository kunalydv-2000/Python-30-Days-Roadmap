# Part 1- __Loops__ in python
---
Loops allow Python to:

- Repeat tasks automatically
- Process large datasets
- Automate calculations
- Handle records one by one

# ```for``` loop
---
Used when you know how many times to repeat.

# Basic Syntax

for variable in sequence:
 code
    

````python
# example 1 - print Numbers
for i in range(5):
    print(i)           # python starts with 0 and ends with n-1
````

````python
# example 2 - start and end
for i in range(1, 6):
    print(i)           # python starts with 1 and ends with n-1
````

````python
# example 3 - step value
for i in range(1, 10, 2):
    print(i)           # python starts with 1 and ends with n-1 and step value is 2
````

````python
# real analytics example
sales = [1200, 1500, 1700, 2000]

for amount in sales:
    print(amount)
````

# Part 3 - Calculations inside loops

````python
# example - Total sales
sales = [1000, 2000, 3000, 4000]

total = 0

for amount in sales:
    total = total + amount

print(total)
````

# Part 4 - `while` loop
---
Used when repetition depends on a condition.

# basic syntax
---
while condition: `code`

````python
# example
count = 1

while count <= 5:
    print(count)
    count += 1
````

# Part 5 - `break`
---
Stops the loop immediately.
Real Use Case:
- Stop search when item found
- Exit menu systems
- Stop invalid operations

````python
# example
for i in range(10):

    if i == 5:
        break

    print(i)
````

# Part 6 - `continue`
---
Skips current iteration.

````python
# example
for i in range(5):

    if i == 2:    # when i is 2, skip the rest of the code and move to the next iteration
        continue

    print(i)
````

# Part 7 - Nested Loops
---
Loop inside another loop.

Real Data Analysis Uses

Nested loops help:

- Compare rows
- Generate combinations
- Process matrices
- Analyze grouped data

````python
# example 1 - sales analysis
sales = [1200, 1500, 1700, 2000]

for sale in sales:
    print(sale * 1.1)    # increase sales by 10%
````

````python
# example 2 - data cleaning
names = ["Kunal", "", "Rahul", ""]

for name in names:

    if name == "":
        continue

    print(name)
````

````python
# example 3 - fraud detection
transactions = [200, 5000, 12000, 300]

for amount in transactions:

    if amount > 10000:
        print("Suspicious Transaction")
````
