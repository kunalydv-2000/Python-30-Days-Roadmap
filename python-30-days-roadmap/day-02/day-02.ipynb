# Part 2- **Data Types**
---
Python stores different types of data

```python
#string(str)
# text data, enclosed in single or double quotes
name = 'kunal'
type(name)  #str
```

```python
#interger(int)
# whole numbers, positive or negative, without decimals, of unlimited length
age = 20
type(age)  #int
```

```python
#float(float)
# numbers with a decimal point or in exponential form
height = 5.8
type(height)  #float
```

```python
#boolean(bool)
# represents one of two values: True or False)
is_student = True
type(is_student)  #bool
```
