# Part 1 - Arthemetic Operators
Arithmetic operators perform calculations.

```python
# Basic Calculations
a = 10
b = 5

print(a + b)
print(a - b)
print(a * b)
print(a / b) # always returns a float
print(a // b) # floor division, returns an integer
```

```python
# Modulus operator (%)
a = 10
b = 3
print(a % b) # Output: 1, because 10 divided by 3 leaves a remainder of 1
```

```python
# power operator (**)
a = 2
b = 3
print(a ** b) # Output: 8, because 2 raised to the power of 3 is 8
```

# Part 2 - Comparision Operators
Comparison operators compare values.

They return:

True
False

```python
age = 20

print(age > 18) # Output: True, because 20 is greater than 18
print(age < 18) # Output: False, because 20 is not less than
```

```python
# = vs == operator
x = 5 # This is an assignment, it assigns the value 5 to the variable x
y = 10 # This is an assignment, it assigns the value 10 to the variable y
print(x == y) # Output: False, because 5 is not equal to 10
print(x == 5) # Output: True, because x is equal to 5
print(x = 5) # This would cause a syntax error, as = is for assignment, not comparison
```

# Part 3 - Conditional Statements (if)
Conditions allow programs to make decisions.

```python
age = 20

if age >= 18:
    print("Eligible to vote")
```

```python
if age >= 18:
    print("Adult") #identation is important in Python, it defines the scope of loops, functions, and other code blocks
```

# Part 4 - if-else
Used when there are 2 possible outcomes.

```python
age = int(input("Enter age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

# Part 5 - if-elif-else
used for multiple conditions

```python
#grade system
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 50:
    print("Grade C")

else:
    print("Fail")
```

# Part 6 - Logical Operators
used to combine conditions

```python
# and, or, not operators
age = 25
salary = 50000

if age > 18 and salary > 30000:
    print("Eligible for loan")

country = "India"

if country == "India" or country == "USA":
    print("Access granted")
    
is_logged_in = False

if not is_logged_in:
    print("Please login")
```
