# Student Grade Calculator

```python
# Student Grade Calculator
print("===== STUDENT RESULT SYSTEM =====")

name = input("Enter student name: ")

marks = float(input("Enter marks: "))

if marks >= 90:
    grade = "A"

elif marks >= 75:
    grade = "B"

elif marks >= 50:
    grade = "C"

else:
    grade = "Fail"

if marks >= 50:
    status = "PASS"
else:
    status = "FAIL"

print("\n===== RESULT =====")

print(f"Student Name: {name}")
print(f"Marks: {marks}")
print(f"Grade: {grade}")
print(f"Status: {status}")
```
